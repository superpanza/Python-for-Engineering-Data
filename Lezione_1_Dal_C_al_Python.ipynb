## Blocco 1.2 — Tipi, strutture dati, control flow

### Variabili: niente dichiarazioni di tipo

In C:
```c
int tensione = 5;
float corrente = 0.01;
char nome[] = "sensore_1";
```

In Python:

````python
tensione = 5          # int (intero)
corrente = 0.01       # float (virgola mobile)
nome = "sensore_1"    # str (stringa)
attivo = True         # bool (booleano)

# Le variabili possono cambiare tipo (typing dinamico):
x = 10
print(f"x = {x}, tipo: {type(x)}")

x = "adesso sono una stringa"
print(f"x = {x}, tipo: {type(x)}")

# In C questo sarebbe un errore di compilazione!
# In Python è legale, ma da usare con cautela.
````

### Operatori: quasi identici al C

````python
# Aritmetica — come in C, con un bonus
a, b = 17, 5          # assegnamento multiplo (non esiste in C!)

print(f"a + b  = {a + b}")
print(f"a - b  = {a - b}")
print(f"a * b  = {a * b}")
print(f"a / b  = {a / b}")      # divisione VERA (float), non troncata!
print(f"a // b = {a // b}")     # divisione intera (come / in C)
print(f"a % b  = {a % b}")      # modulo
print(f"a ** b = {a ** b}")     # potenza (in C serviva pow())
````

### Stringhe e f-string: il `printf()` evoluto

Dimenticate `printf("%s ha tensione %.2f V", nome, v)`.  
Le f-string di Python sono più leggibili e potenti:

````python
nome_sensore = "ADS1115"
tensione = 3.2956
canale = 2
campioni = 1024

# f-string: metti f prima della stringa, e le variabili tra {}
print(f"Sensore: {nome_sensore}")
print(f"CH{canale}: {tensione:.2f} V")        # 2 decimali
print(f"CH{canale}: {tensione:.4e} V")        # notazione scientifica
print(f"Campioni: {campioni:>8d}")            # allineato a destra, 8 caratteri
print(f"Campioni: {campioni:08d}")            # zero-padded
print(f"Guadagno: {20 * 3.14159:.1f} dB")    # espressioni dentro {}
````

### Liste: l'array dinamico che in C non avevi

````python
# Creazione
misure = [1.2, 3.4, 5.6, 7.8]
print(f"Misure: {misure}")
print(f"Lunghezza: {len(misure)}")

# Accesso per indice (come in C, parte da 0)
print(f"Prima misura: {misure[0]}")
print(f"Ultima misura: {misure[-1]}")   # indice negativo: conta dalla fine!

# Slicing: estrarre sottoliste (non esiste in C!)
print(f"Prime 2: {misure[:2]}")    # da 0 a 1
print(f"Ultime 2: {misure[-2:]}")  # ultime 2
print(f"Dal 1 al 2: {misure[1:3]}")  # indici 1, 2 (il 3 è escluso!)
````

````python
# Le liste sono DINAMICHE (niente malloc/realloc)
campioni = []                     # lista vuota
for i in range(5):
    campioni.append(i * 0.1)      # aggiungi alla fine
print(f"Campioni: {campioni}")

# Operazioni utili
campioni.insert(0, -0.1)         # inserisci alla posizione 0
print(f"Dopo insert: {campioni}")

ultimo = campioni.pop()           # rimuovi e restituisci l'ultimo
print(f"Pop: {ultimo}, Lista: {campioni}")

# Le liste possono contenere tipi misti (ma meglio evitare)
misto = [1, "due", 3.0, True]
print(f"Lista mista: {misto}")
````

### Tuple: come le liste, ma immutabili

Utili per coordinate, parametri fissi, valori di ritorno multipli.

````python
# Tuple: parentesi tonde, non si possono modificare
punto = (3.5, 7.2)
print(f"x = {punto[0]}, y = {punto[1]}")

# Unpacking: estrarre i valori in variabili separate
x, y = punto
print(f"x = {x}, y = {y}")

# Uso tipico: parametri di configurazione che non devono cambiare
BAUD_RATES = (9600, 19200, 38400, 57600, 115200)
print(f"Baud rates disponibili: {BAUD_RATES}")

# Se provi a modificarla:
# punto[0] = 5.0   # ← TypeError! Le tuple sono immutabili
````

### Dizionari: la struct del Python

In C:
```c
struct Sensore {
    char nome[32];
    float gain;
    float offset;
    int canale;
};
struct Sensore s = {"ADS1115", 10.0, 0.3, 2};
```

In Python:

````python
# Dizionario: coppie chiave-valore
sensore = {
    "nome": "ADS1115",
    "gain": 10.0,
    "offset": 0.3,
    "canale": 2,
    "attivo": True
}

# Accesso per chiave (come i campi di una struct)
print(f"Sensore: {sensore['nome']}")
print(f"Guadagno: {sensore['gain']}")

# Aggiungere/modificare campi (la struct cresce dinamicamente!)
sensore["fs"] = 860  # samples/sec
sensore["gain"] = 20.0
print(f"Config aggiornata: {sensore}")

# Iterare sulle chiavi
for chiave, valore in sensore.items():
    print(f"  {chiave}: {valore}")
````

````python
# Caso d'uso reale: configurazione di un esperimento
config_esperimento = {
    "nome": "sweep_frequenza_v2",
    "strumento": "Rigol DSA815",
    "freq_start": 1e6,       # 1 MHz
    "freq_stop": 100e6,      # 100 MHz
    "punti": 601,
    "unita": "dBm"
}

# Calcolo derivato
span = config_esperimento["freq_stop"] - config_esperimento["freq_start"]
step = span / (config_esperimento["punti"] - 1)
print(f"Span: {span/1e6:.0f} MHz")
print(f"Step: {step/1e3:.1f} kHz")
````

### Control flow: stessa logica, sintassi diversa

La differenza principale: in Python **l'indentazione È la sintassi**.
Niente `{ }`, niente `;`.

````python
# if / elif / else
tensione = 4.2

if tensione > 5.0:
    print("SOVRATENSIONE!")
elif tensione > 3.0:
    print("Tensione nominale")
elif tensione > 0:
    print("Tensione bassa")
else:
    print("Nessuna tensione")
````

## Blocco 1.3 — Funzioni, moduli, comprehension

### Definire funzioni: `def` al posto della signature C

In C:
```c
float volt_to_dbm(float v_rms, float z) {
    float p = (v_rms * v_rms) / z;
    return 10.0 * log10(p / 0.001);
}
```

In Python:

````python
import math

def volt_to_dbm(v_rms, z=50):
    """Converte tensione RMS in dBm.

    Args:
        v_rms: tensione RMS in Volt
        z: impedenza in Ohm (default: 50)

    Returns:
        potenza in dBm
    """
    p_watt = (v_rms ** 2) / z
    return 10 * math.log10(p_watt / 1e-3)

# Uso con impedenza di default (50 Ω)
print(f"1V RMS su 50Ω = {volt_to_dbm(1.0):.2f} dBm")

# Uso con impedenza specificata
print(f"1V RMS su 75Ω = {volt_to_dbm(1.0, z=75):.2f} dBm")

# z=50 è un argomento con default: se non lo passi, vale 50
# In C non esiste! (in C++ sì)
````

### Valori di ritorno multipli: niente puntatori!

In C per restituire più valori servono puntatori:
```c
void calc_stats(float *data, int n, float *mean, float *std);
```

In Python è naturale:

````python
def statistiche(dati):
    """Calcola media e deviazione standard di una lista."""
    n = len(dati)
    media = sum(dati) / n
    varianza = sum((x - media) ** 2 for x in dati) / n
    dev_std = math.sqrt(varianza)
    return media, dev_std   # restituisce una TUPLA

# Unpacking del risultato
misure = [3.28, 3.31, 3.29, 3.33, 3.27, 3.30, 3.32]
media, std = statistiche(misure)
print(f"V = {media:.3f} ± {std:.3f} V")

# Equivalente a:
risultato = statistiche(misure)
print(f"Tipo del risultato: {type(risultato)}")
print(f"Media: {risultato[0]:.3f}, Std: {risultato[1]:.3f}")
````

### List comprehension: il superpotere di Python

Un `for` compatto che crea una lista in una sola riga.
Non esiste in C, ed è una delle feature più usate di Python.

````python
# Modo C-like (funziona, ma non è pythonico):
risultati = []
for f in [100, 1000, 10000]:
    periodo = 1 / f
    risultati.append(periodo)
print(f"Periodi (loop): {risultati}")

# Modo Python (list comprehension):
periodi = [1/f for f in [100, 1000, 10000]]
print(f"Periodi (comp): {periodi}")

# È la stessa cosa, ma più compatta e leggibile (con l'abitudine)
````

````python
# 1. Convertire una lista di tensioni in dBm
v_list = [0.001, 0.01, 0.1, 1.0, 10.0]
dbm_list = [volt_to_dbm(v) for v in v_list]
for v, dbm in zip(v_list, dbm_list):
    print(f"  {v:>6.3f} V → {dbm:>8.2f} dBm")
````

````python
# 2. Generare frequenze logaritmiche (come uno sweep)
freq_log = [10 ** (i/10) for i in range(10, 51)]  # da 10 Hz a ~100 kHz
print(f"Prime 5 frequenze: {[f'{f:.0f}' for f in freq_log[:5]]} Hz")
print(f"Ultime 5 frequenze: {[f'{f:.0f}' for f in freq_log[-5:]]} Hz")
print(f"Totale punti: {len(freq_log)}")
````

````python
# 3. Comprehension con filtro (if)
misure_raw = [3.3, 999.9, 3.2, 3.4, -1.0, 3.3, 999.9, 3.1]

# Filtra i valori fuori range (outlier/errori del sensore)
misure_ok = [m for m in misure_raw if 0 < m < 10]
print(f"Raw:    {misure_raw}")
print(f"Pulite: {misure_ok}")
print(f"Media:  {sum(misure_ok)/len(misure_ok):.2f} V")
````

### Gestione errori: `try/except`

In C un errore spesso = segmentation fault.
In Python possiamo catturare e gestire gli errori:

````python
def divisore_tensione(v_in, r1, r2):
    """Calcola V_out di un partitore resistivo."""
    try:
        v_out = v_in * r2 / (r1 + r2)
        return v_out
    except ZeroDivisionError:
        print("Errore: R1 + R2 = 0!")
        return None

# Funziona:
print(f"Vout = {divisore_tensione(5.0, 1000, 1000):.2f} V")

# Gestisce l'errore senza crashare:
print(f"Vout = {divisore_tensione(5.0, 0, 0)}")

# Gestione di file non trovato:
try:
    with open("dati_inesistenti.csv", "r") as f:
        dati = f.read()
except FileNotFoundError:
    print("File non trovato! Controlla il percorso.")
````

````python
# ═══════════════════════════════════════════
# SOLUZIONI
# ═══════════════════════════════════════════

import math

# --- Parte 1: Funzioni di conversione ---

def dbm_to_watt(dbm):
    """Converte dBm in Watt."""
    return 10 ** (dbm / 10) * 1e-3

def watt_to_dbm(w):
    """Converte Watt in dBm."""
    return 10 * math.log10(w / 1e-3)

def dbv_to_volt(dbv):
    """Converte dBV in Volt."""
    return 10 ** (dbv / 20)

def volt_to_dbv(v):
    """Converte Volt in dBV."""
    return 20 * math.log10(v)

def celsius_to_fahrenheit(c):
    return c * 9/5 + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

def rad_to_deg(rad):
    return rad * 180 / math.pi

def deg_to_rad(deg):
    return deg * math.pi / 180

# --- Parte 2: Test ---

print("=== Test conversioni ===")
print(f"0 dBm  = {dbm_to_watt(0)*1000:.1f} mW  (atteso: 1.0)")
print(f"30 dBm = {dbm_to_watt(30):.1f} W   (atteso: 1.0)")
print(f"1 mW   = {watt_to_dbm(0.001):.1f} dBm (atteso: 0.0)")
print(f"0 dBV  = {dbv_to_volt(0):.1f} V   (atteso: 1.0)")
print(f"100 °C = {celsius_to_fahrenheit(100):.0f} °F  (atteso: 212)")
print(f"π rad  = {rad_to_deg(math.pi):.0f}°    (atteso: 180)")

# --- Parte 3: Tabella ---

print("\n=== Tabella dBm → Watt ===")
dbm_values = [-30, -20, -10, 0, 10, 20, 30]
for dbm in dbm_values:
    w = dbm_to_watt(dbm)
    if w < 1e-3:
        print(f"{dbm:>4d} dBm → {w*1e6:>8.3f} µW")
    else:
        print(f"{dbm:>4d} dBm → {w*1e3:>8.3f} mW")
````
